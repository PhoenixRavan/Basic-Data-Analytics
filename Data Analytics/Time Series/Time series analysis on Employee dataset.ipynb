## 8)Visualize the seasonality of multiple time series and the correlation between each time series in the dataset. 

Seasonality on the basis of month and quarterly

```python
df_month = df.resample("M").mean()
df_quar = df.resample("Q").mean()


for str in my_list:
    decomposition = sm.tsa.seasonal_decompose(df_month[str], model='Additive').seasonal
    decomp = sm.tsa.seasonal_decompose(df_quar[str], model='Additive').seasonal
    #decom = sm.tsa.seasonal_decompose(df_fort[my_list[i]], model='Additive').seasonal
    fig = decomposition.plot()
    big = decomp.plot()
    #wig = decom.plot()
    plt.title(str)
    plt.show()
```

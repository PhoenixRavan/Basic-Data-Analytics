```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
df = pd.read_csv("groceries.csv")
df
```

```python
records=list()

for i in range(0,9835):
     records.append([str(df.values[i,j]) for j in range(0,32)])
records
```

## Preparing the data for Association Rule Mining

```python
te = TransactionEncoder()
arr=te.fit(records).transform(records)
df = pd.DataFrame(arr, columns=te.columns_)



del df['nan']
df.dropna()
df.head(11)
df=df.iloc[1:]
df.head(11)
```

## 9) No. of transactions and itemsets in the Dataframe

```python
df.shape
```

There are 9835 Transactions and 169 itemsets

## 12) Find top selling items with minimum support of 2%

```python
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets[ (frequent_itemsets['length'] == 1) & (frequent_itemsets['support'] >= 0.02) ].sort_values(by='support', ascending=False)
```

## 13) Find all frequent itemsets with minimum support of 5%

```python
frequent_itemsets[(frequent_itemsets['length'] != 1)&(frequent_itemsets['support'] >= 0.05)].sort_values(by='support', ascending=False)
```

## 14) Find all frequent itemsets of length 2 with minimum support of 2%

```python
frequent_itemsets[(frequent_itemsets['length'] == 2)&(frequent_itemsets['support'] >= 0.02) ].sort_values(by='support', ascending=False)
```

## 15) Find the top 10 association ruleswith minimum support of 2%, sorted by confidence in descending order

```python
association_rules(frequent_itemsets, metric="support", min_threshold=0.02).sort_values(by='confidence', ascending=False)
```

## 16) Find association rules with minimum support of 2% and lift of more than 1

```python
df=association_rules(frequent_itemsets, metric="support", min_threshold=0.02)
df[df['lift'] > 1 ]
```
